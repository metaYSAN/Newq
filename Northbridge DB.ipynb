First select a workspace and the `information_schema` database from the drop-down menu at the top of this notebook.

### SQL Queries for Website Features

#### 1. Search bar to look for companies 

```python
%%sql
SELECT company_name, company_location, contact_person, contact_email, contact_phone 
FROM employers 
WHERE company_name LIKE '%greentech%' 
AND time_out = '99991231'
```

#### 2. Recommended candidates by universities

```python
%%sql
SELECT students.student_id, name, email, phone, visa_type
FROM students
JOIN student_universities
    ON students.student_id = student_universities.student_id 
JOIN visa_statuses 
    ON students.visa_status_id = visa_statuses.visa_status_id
WHERE students.time_out = '99991231' 
    AND visa_statuses.time_out = '99991231'
    AND student_universities.time_out = '99991231'
    AND student_universities.university_id = '1'
```

#### 3. Find jobs sponsoring VISA status

```python
%%sql
SELECT title, description, location, posting_date, visa_sponsorship_list, application_deadline, company_name, contact_person, contact_email, contact_phone
FROM jobs 
JOIN employers 
    ON jobs.employer_id = employers.employer_id 
WHERE jobs.time_out = '99991231' 
    AND employers.time_out = '99991231' 
    AND visa_sponsorship_list like '%1%'
```

#### 4. Get available job opportunities based on profiles (skills)

```python
%%sql
SELECT title, description, location, posting_date, application_deadline, company_name, contact_person, contact_email, contact_phone
FROM jobs 
JOIN employers 
    ON jobs.employer_id = employers.employer_id 
WHERE jobs.time_out = '99991231' 
    AND employers.time_out = '99991231' 
    AND (description LIKE '%development%' OR title LIKE '%development%')
```

#### 5. Application tracker (track application status)

```python
%%sql
SELECT application_date, status, applications.time_in AS update_time 
FROM applications JOIN students 
    ON applications.student_id = students.student_id 
WHERE students.student_id = 1 
    AND application_id = 1
    AND applications.time_out = '99991231' 
    AND students.time_out = '99991231'
```

#### 6. Search candidates by degree/major

```python
%%sql
SELECT name, email, phone, resume, background, visa_status_id, university_name, degree_level, first_major, second_major
FROM students JOIN student_universities 
    ON students.student_id = student_universities.student_id 
JOIN universities ON universities.university_id = student_universities.university_id
WHERE students.time_out = '99991231' 
    AND students.time_out = '99991231' 
    AND universities.time_out = '99991231' 
    AND student_universities.is_enrolled is TRUE
    AND (first_major LIKE '%Computer Science%' 
        OR first_major LIKE '%Artificial Intelligence%' 
        OR second_major LIKE '%Computer Science%' 
        OR second_major LIKE '%Artificial Intelligence%')
    AND (degree_level LIKE 'Bachelor' OR degree_level LIKE 'Master')
```

#### 7. Search candidates by skills

```python
%%sql
SELECT name, email, phone, resume, background, visa_status_id, university_name, degree_level, first_major, second_major
FROM students JOIN student_universities 
    ON students.student_id = student_universities.student_id 
JOIN universities ON universities.university_id = student_universities.university_id 
JOIN student_skills ON students.student_id = student_skills.student_id 
JOIN skills ON student_skills.skill_id = skills.skill_id 
WHERE (skills.skill_name LIKE '%Machine Learning%'  OR skills.skill_name LIKE '%Python%')
    AND students.time_out = '99991231' 
    AND universities.time_out = '99991231' 
    AND student_skills.time_out = '99991231' 
    AND skills.time_out = '99991231'
    AND is_enrolled IS TRUE
GROUP BY students.student_id
```

#### 8. Search jobs based on location(s)

```python
%%sql
SELECT title, description, location, posting_date, application_deadline, company_name, contact_person, contact_email, contact_phone
FROM jobs 
JOIN employers 
    ON jobs.employer_id = employers.employer_id 
WHERE jobs.time_out = '99991231' 
    AND employers.time_out = '99991231' 
    AND location LIKE '%, CA%'
```

```python
def create_id_for_new_entry(conn, table_name): 
    table_id_mapping = {
        'applications': 'application_id', 
        'employer_ratings': 'rating_id', 
        'employers': 'employer_id', 
        'jobs': 'job_id', 
        'skills': 'skill_id', 
        'students': 'student_id', 
        'universities': 'university_id', 
        'visa_statuses': 'visa_status_id'
    }
    query = f"SELECT {table_id_mapping[table_name]} from {table_name} order by student_id + 0 DESC limit 1"
    student_id = conn.execute(sa.text(query)).mappings().all()[0]['student_id']
    return str(int(student_id) + 1)

create_id_for_new_entry(conn, 'students')
```

#### 1. Search bar to look for companies 

```python
def get_companies_by_name(conn, company_name): 
    query = f"""
        SELECT company_name, company_location, contact_person, contact_email, contact_phone 
        FROM employers 
        WHERE company_name LIKE '%{company_name}%' 
            AND time_out = '99991231' 
    """
    result = conn.execute(sa.text(query))
    return {
        'companies': result.mappings().all()
    }

get_companies_by_name(conn, 'greentech')
```

#### 2. Recommended candidates by universities

```python
def get_candidates_by_universities(conn, university_id): 
    query = f"""
        SELECT students.student_id, name, email, phone, visa_type
        FROM students
        JOIN student_universities
            ON students.student_id = student_universities.student_id 
        JOIN visa_statuses 
            ON students.visa_status_id = visa_statuses.visa_status_id
        WHERE students.time_out = '99991231' 
            AND visa_statuses.time_out = '99991231'
            AND student_universities.time_out = '99991231'
            AND student_universities.university_id = {university_id}
    """
    result = conn.execute(sa.text(query))
    return {
        'candidates': result.mappings().all()
    }

get_candidates_by_universities(conn, 1)
```

#### 3. Find jobs sponsoring VISA status

```python
def get_jobs_by_visa_status(conn, visa_status_id): 
    query = f"""
        SELECT title, description, location, posting_date, visa_sponsorship_list, application_deadline, company_name, contact_person, contact_email, contact_phone
        FROM jobs 
        JOIN employers 
            ON jobs.employer_id = employers.employer_id 
        WHERE jobs.time_out = '99991231' 
            AND employers.time_out = '99991231' 
            AND visa_sponsorship_list like '%{visa_status_id}%'
    """
    result = conn.execute(sa.text(query))
    return {
        'jobs': result.mappings().all()
    }

get_jobs_by_visa_status(conn, 1)
```

#### 4. Get available job opportunities based on profiles (skills)

```python
def get_jobs_by_candidate_skills(conn, skills): 
    # skills is a list of str where each element is a user skill input e.g. skills = ['software', 'developer', 'engineer']
    skills_subquery = ' OR '.join([f"description LIKE '%{skill}%' OR title LIKE '%{skill}%'" for skill in skills])
    query = f"""
        SELECT title, description, location, posting_date, application_deadline, company_name, contact_person, contact_email, contact_phone
        FROM jobs 
        JOIN employers 
            ON jobs.employer_id = employers.employer_id 
        WHERE jobs.time_out = '99991231' 
            AND employers.time_out = '99991231' 
            AND ({skills_subquery})
        """
    result = conn.execute(sa.text(query))
    return {
        'jobs': result.mappings().all()
    }

get_jobs_by_candidate_skills(conn, ['software', 'developer', 'engineer'])
```

#### 5. Application tracker (track application status)

```python
def get_application_status(conn, student_id):
    query = f"""SELECT application_date, status, applications.time_in AS update_time 
        FROM applications JOIN students 
            ON applications.student_id = students.student_id 
        WHERE students.student_id = {student_id}
            AND applications.time_out = '99991231' 
            AND students.time_out = '99991231'
    """
    result = conn.execute(sa.text(query))
    return {
        'application_statuses': result.mappings().all()
    }

get_application_status(conn, 1)
```

#### 6. Search candidates by degree/major

```python
def get_candidates_by_majors(conn, majors): 
    majors_subquery = ' OR '.join([f"first_major LIKE '%{major}%' OR second_major LIKE '%{major}%'" for major in majors])
    query = f"""
        SELECT name, email, phone, resume, background, visa_status_id, university_name, degree_level, first_major, second_major
        FROM students JOIN student_universities 
            ON students.student_id = student_universities.student_id 
        JOIN universities ON universities.university_id = student_universities.university_id
        WHERE students.time_out = '99991231' 
            AND students.time_out = '99991231' 
            AND universities.time_out = '99991231' 
            AND student_universities.is_enrolled is TRUE
            AND ({majors_subquery})
    """
    result = conn.execute(sa.text(query))
    return {
        'candidates': result.mappings().all()
    }

get_candidates_by_majors(conn, ['Computer Science', 'Artificial Intelligence', 'Mathematics'])
```

#### 7. Search candidates by skills

```python
def get_candidates_by_skills(conn, skills): 
    skills_subquery = ' OR '.join([f"skills.skill_name LIKE '%{skill}%'" for skill in skills])
    query = f"""
        SELECT name, email, phone, resume, background, visa_status_id, university_name, degree_level, first_major, second_major
        FROM students JOIN student_universities 
            ON students.student_id = student_universities.student_id 
        JOIN universities ON universities.university_id = student_universities.university_id 
        JOIN student_skills ON students.student_id = student_skills.student_id 
        JOIN skills ON student_skills.skill_id = skills.skill_id 
        WHERE ({skills_subquery})
            AND students.time_out = '99991231' 
            AND universities.time_out = '99991231' 
            AND student_skills.time_out = '99991231' 
            AND skills.time_out = '99991231'
            AND is_enrolled IS TRUE
        GROUP BY students.student_id
    """
    result = conn.execute(sa.text(query))
    return {
        'candidates': result.mappings().all()
    }

get_candidates_by_skills(conn, ['Python', 'Machine Learning', 'SQL'])
```

#### 8. Search jobs based on location(s)

```python
def get_jobs_by_locations(conn, locations, accept_remote=True): 
    # location is a list of states
    if accept_remote:
        locations += ['remote']
    locations_subquery = ' OR '.join([f"location LIKE '%, {location}%'" for location in locations])
    query = f"""
        SELECT title, description, location, posting_date, application_deadline, company_name, contact_person, contact_email, contact_phone
        FROM jobs 
        JOIN employers 
            ON jobs.employer_id = employers.employer_id 
        WHERE jobs.time_out = '99991231' 
            AND employers.time_out = '99991231' 
            AND ({locations_subquery})
    """
    result = conn.execute(sa.text(query))
    return {
        'jobs': result.mappings().all()
    }

get_jobs_by_locations(conn, ['CA', 'NY', 'TX'])
```
